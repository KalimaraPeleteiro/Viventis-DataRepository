```python
import pandas as pd
```

```python
dataset = pd.read_csv("../data/tigganeha4-diabetes2019.csv")
dataset
```

### Tratamento

```python
dataset.Diabetic.value_counts()
```

```python
dataset.info()
```

```python
dataset.BMI.fillna(dataset.BMI.median(), inplace=True)
dataset.Pregancies.fillna(0.0, inplace=True)
```

```python
dataset.info()
```

```python
dataset.dropna(inplace=True, subset=['Diabetic'])
```

```python
dataset.info()
```

```python
dataset.Pdiabetes.value_counts()
```

```python
dataset.Pdiabetes.fillna(0, inplace=True)
```

```python
dataset.info()
```

### Mudando Tipos

```python
dataset.Gender.replace(['Male', 'Female'], [0, 1], inplace=True)
```

```python
dataset.Age.replace(['less than 40', '40-49', '50-59', '60 or older'], [0, 1, 2, 3], inplace=True)
```

```python
dataset.Family_Diabetes.replace(['yes', 'no'], [1, 0], inplace=True)
dataset.Smoking.replace(['yes', 'no'], [1, 0], inplace=True)
dataset.Alcohol.replace(['yes', 'no'], [1, 0], inplace=True)
dataset.highBP.replace(['yes', 'no'], [1, 0], inplace=True)
```

```python
dataset.PhysicallyActive.replace(['less than half an hr', 'more than half an hr', 'one hr or more', 'none'], [1, 2, 3, 0], inplace=True)
```

```python
dataset.RegularMedicine.replace(['yes', 'no', 'o'], [1, 0, 0], inplace=True)
```

```python
dataset.JunkFood.replace(['occasionally', 'often', 'very often', 'always'], [0, 1, 2, 3], inplace=True)
```

```python
dataset.BPLevel.replace(['low', 'Low', 'normal', 'normal ', 'high', 'High'], [0, 0, 1, 1, 2, 2], inplace=True)
```

```python
dataset.info()
```

```python
dataset.Stress.replace(['not at all', 'sometimes', 'very often', 'always'], [0, 1, 2, 3], inplace=True)
```

```python
dataset.Pdiabetes.replace(['0', 'no', 'yes'], [0, 0, 1], inplace=True)
```

```python
dataset.UriationFreq.replace(['not much', 'quite often'], [0, 1], inplace=True)
```

```python
dataset.Diabetic.replace(['no', ' no', 'yes'], [0, 0, 1], inplace=True)
```

```python
dataset.info()
```

### Preparação

```python
y = dataset["Diabetic"]
x = dataset.drop("Diabetic", axis=1)

x
```

## Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

x_scaled = pd.DataFrame(x_scaled, columns=x.columns)
x_scaled
```

```python
scaled_dataset = x_scaled.copy()
scaled_dataset["Diabetic"] = y

scaled_dataset
```

### Correlação

```python
correlation = scaled_dataset.corr(method='spearman')
correlation["Diabetic"].sort_values(ascending=False, key=abs)
```

```python
correlation = scaled_dataset.corr(method='kendall')
correlation["Diabetic"].sort_values(ascending=False, key=abs)
```

```python
correlation = scaled_dataset.corr(method='pearson')
correlation["Diabetic"].sort_values(ascending=False, key=abs)
```

### Importância das Variáveis

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import numpy as np

model = RandomForestClassifier(n_estimators=100)
model.fit(x_scaled, y)

importances = model.feature_importances_

indices = np.argsort(importances)[::-1]  
sorted_importances = importances[indices]
sorted_features = x_scaled.columns[indices]

plt.figure(figsize=(10, 6))
plt.barh(sorted_features, sorted_importances)
plt.xlabel('Importância das Variáveis')
plt.ylabel('Variável')
plt.title('Modelo: RandomForest')
plt.show()
```

```python
import xgboost as xgb
import matplotlib.pyplot as plt
import numpy as np

model = xgb.XGBClassifier(n_estimators=100)
model.fit(x_scaled, y)

importances = model.feature_importances_

indices = np.argsort(importances)[::-1]  
sorted_importances = importances[indices]
sorted_features = x_scaled.columns[indices]

plt.figure(figsize=(10, 6))
plt.barh(sorted_features, sorted_importances)
plt.xlabel('Importância das Variáveis')
plt.ylabel('Variável')
plt.title('Modelo: XGBoost')
plt.show()
```

## Treinamento Inicial

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb

SEED = 42

X_train, X_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, stratify=y, random_state=SEED)

dt_model = DecisionTreeClassifier(random_state=SEED)
dt_model.fit(X_train, y_train)
dt_preds = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_preds)
dt_precision = precision_score(y_test, dt_preds, average='binary')  
dt_recall = recall_score(y_test, dt_preds, average='binary')  


rf_model = RandomForestClassifier(n_estimators=100, random_state=SEED)
rf_model.fit(X_train, y_train)
rf_preds = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_preds)
rf_precision = precision_score(y_test, rf_preds, average='binary')  
rf_recall = recall_score(y_test, rf_preds, average='binary')  


xgb_model = xgb.XGBClassifier(n_estimators=100, random_state=SEED)
xgb_model.fit(X_train, y_train)
xgb_preds = xgb_model.predict(X_test)
xgb_accuracy = accuracy_score(y_test, xgb_preds)
xgb_precision = precision_score(y_test, xgb_preds, average='binary')  
xgb_recall = recall_score(y_test, xgb_preds, average='binary')  

print("Acurácia: Número Total de Casos Corretos.")
print("Recall: De todos os casos com diabetes, quantos o modelo acertou?")
print("Precisão: De todos os casos que o modelos disse ser diabetes, quantos realmente eram?")

print(f"\nDecision Tree - Acurácia: {dt_accuracy:.4f}, Precisão: {dt_precision:.4f}, Recall: {dt_recall:.4f}")
print(f"\nRandom Forest - Acurácia: {rf_accuracy:.4f}, Precisão: {rf_precision:.4f}, Recall: {rf_recall:.4f}")
print(f"\nXGBoost - Acurácia: {xgb_accuracy:.4f}, Precisão: {xgb_precision:.4f}, Recall: {xgb_recall:.4f}")
```
